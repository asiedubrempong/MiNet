## Data

```python
import os                                                                                                             

def list_files(dir):                                                                                                  
    r = []                                                                                                            
    subdirs = [x[0] for x in os.walk(dir)]                                                                            
    for subdir in subdirs:                                                                                            
        files = next(os.walk(subdir))[2]                                                                             
        if (len(files) > 0):                                                                                          
            for file in files:                                                                                        
                r.append(subdir + "/" + file)                                                                         
    return r
```

```python
name = list_files('data/')
label = [path.split('/')[1] for path in name]
name_label_dict = {'name': name, 'label': label}
df = pd.DataFrame(name_label_dict, columns=['name', 'label'])
df.head()
```

```python
skf = StratifiedKFold(n_splits=5)
```

```python
test_splits = []
for train, test in skf.split(name, label):
    print("%s %s" % (train, test))
    test_splits.append(test)
```

```python
test_splits
```

```python
with open('splits.pkl', 'wb') as f:
    pickle.dump(test_splits, f)
```

```python
del name
del label
del name_label_dict
del df
```

```python
path = Path('')
tfms = get_transforms(flip_vert=True)
```

```python
src = (ImageList.from_df(df, path)
       .split_by_idx(test_splits[0])
       .label_from_df())
```

```python
data = (src.transform(tfms, size=224)
        .databunch(bs=32).normalize(imagenet_stats))
```

## Model

```python
learn = cnn_learner(data, models.densenet121, metrics=error_rate)
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
lr = 1e-2
learn.fit_one_cycle(5, slice(lr))
```

```python
learn.save('stage-1-dn121')
```

### Unfreeze and finetune

```python
learn.unfreeze()
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, slice(1e-4, lr/10))
```

```python
learn.save('fold1-dnfull')
```

```python
learn.export('fold1-dnfull.pkl')
```

## Model Interpretation

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_top_losses(4)
```

```python
interp.plot_confusion_matrix(dpi=130)
```
